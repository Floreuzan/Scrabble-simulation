```python
import random
import matplotlib.pyplot as plt
%matplotlib inline

import ps2
import ps3a
import seaborn as sns
```

```python
def plot_mc_hand(hand, N, n):
    
    _, mc_scores = ps3a.play_n_mc_hand(hand, N, n)
    plt.figure()
    #plt.hist(mc_scores, facecolor='blue', alpha=0.5)
    sns.distplot(mc_scores)
    
    
    # naming the x axis 
    plt.xlabel('x - axis') 
    # naming the y axis 
    plt.ylabel('y - axis') 
    z='N=', str(N), 'n=', str(n)
    # giving a title to my graph 
    plt.title(z)
    
```

```python
hand = ps2.get_frequency_dict('helloworld')
random.seed(1)
for n in (10, 100):
    for N in (1, 10, 100):
        plot_mc_hand(hand, N, n)
        
```
